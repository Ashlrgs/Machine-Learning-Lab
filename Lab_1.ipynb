```python
#Tukey’s rule to detect outliers
Q1 = df1['age'].quantile(0.25)
Q3 = df1['age'].quantile(0.75)
IQR =Q3-Q1
```

```python
IQR
```

```python
lower_lim=Q1-1.5*IQR
upper_lim=Q3+1.5*IQR
```

```python
lower_lim
```

```python
upper_lim
```

```python
#dropping outlierss
outliers_low=(df1['age']<lower_lim)
```

```python
outliers_up=(df1['age']>upper_lim)
```

```python
len(df1['age'])-(len(df1['age'][outliers_low])+len(df1['age'][outliers_up]))
```

```python
df1['age'][(outliers_low | outliers_up)]
```

```python
df1['age'][~(outliers_low | outliers_up)]
```

```python
df1
```

```python
df1 = df1[~(outliers_low | outliers_up)]
```
